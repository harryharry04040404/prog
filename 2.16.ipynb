### 2.16  Nombres ambiciosos

```python
def divisors(num):
    sumDiv = 1
    for i in range(num):
        if i > 0:
            if (num % i == 0):
                sumDiv += i
    return sumDiv
```

```python
def ambicios(num):  
    b = False
    i = 0
    while i < 250:
        numSum = divisors(num)
        numSum2 = divisors(numSum)
        print(num, numSum, numSum2)
        if (num == numSum) or (numSum == numSum2):
            b = True  
        i += 1
    return b
```

```python
assert ambicios(25) == True  
assert ambicios(100) == False 
assert ambicios(143) == True 
```
